```
import cv2
import numpy as np
import os
import glob

from matplotlib import pyplot as plt
from skimage import io, feature
from scipy import ndimage
from google.colab.patches import cv2_imshow

  
# Function to extract frames 
def FrameCapture(path): 
      
    # Path to video file 
    vidObj = cv2.VideoCapture(path) 
    # Used as counter variable 
    count = 0
    # checks whether frames were extracted 
    success = 1
  
    while success: 
        # vidObj object calls read 
        # function extract frames 
        success, image = vidObj.read() 
        if image is None:
          break
        #Convert to greyscale
        greyFrame = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        # Saves the frames with frame-count 
        cv2.imwrite("/content/drive/My Drive/HW2Cs188/save/frame%d.jpg" % count, image) 
        cv2.imwrite("/content/drive/My Drive/HW2Cs188/saveGrey/greyFrame%d.jpg" % count, greyFrame)
        count += 1  
    # Calling the function 
    FrameCapture("/content/drive/My Drive/HW2Cs188/vid1.mp4") 




#Test one frame for the correlation coefficient Matrix
img_rgb = cv2.imread('/content/drive/My Drive/HW2Cs188/frame1.jpg')
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
template = cv2.imread('/content/drive/My Drive/HW2Cs188/cupTemplate.jpg',0)
w, h = template.shape[::-1]

res = cv2.matchTemplate(img_gray,template,cv2.TM_CCOEFF_NORMED)
threshold = 0.8
loc = np.where( res >= threshold)
for pt in zip(*loc[::-1]):
    cv2.rectangle(img_gray, pt, (pt[0] + w, pt[1] + h), (0,0,255), 2)

cv2.imwrite('/content/drive/My Drive/HW2Cs188/res.jpg', img_gray)

#plt.subplot(121),
#plt.xlabel('Pixel Location in X Direction')
#plt.ylabel('Pixel Location in Y Direction')
io.imshow(res,cmap = 'gray')
plt.show()


#To Plot on Google Colab

%matplotlib inline

#Load the Data
def load_image (folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder, filename))
        if img is not None:
            img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            images.append(img_gray)
    return images

frames = []
for img in glob.glob("/content/drive/My Drive/HW2Cs188/test/*.jpg"):
    fr = cv2.imread(img)
    fr_gray = cv2.cvtColor(fr, cv2.COLOR_BGR2GRAY)
    frames.append(fr_gray)

template = cv2.imread('/content/drive/My Drive/HW2Cs188/template_box.jpg', 0)
w,h = template.shape[::-1]

max_coef = []
for frame in frames:
    res = cv2.matchTemplate(frame, template, cv2.TM_CCOEFF_NORMED)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    max_coef.append(max_loc)

xcor = []
ycor = []
for x, y in max_coef:
    xcor.append(x)
    ycor.append(y)
print(xcor)
print(ycor)

diffx = []
diffy = []
for x in xcor:
    diffx.append(x - xcor[0])
for y in ycor:
    diffy.append(y - ycor[0])
print(diffx)
print(diffy)

#T = []
#for i in range(len(diffx)):
#    T.append(np.float32([[1, 0, -1 * diffx[i]], [0, 1, -1 * diffy[i]]]))
#print(T)

T = np.float32([[1, 0, -1 * diffx[75]], [0, 1, -1 * diffy[75]]])

image = cv2.imread('/content/drive/My Drive/HW2Cs188/test/frame75.jpg')
height, width = image.shape[:2]
img_translation = cv2.warpAffine(image, T, (width, height))
#cv2.imshow('Translation', img_translation)
cv2_imshow(cv2.resize(img_translation, (1920, 1080)))
```
